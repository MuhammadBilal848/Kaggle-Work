```python
dfi = pd.read_csv('Final - Notebook Tier Complete Data.csv')
```

```python
df = dfi['Country']
```

```python
lst = []
for a in df:
    if type(a) == float:
        cty , state , cntry = 'NA' , 'NA' , 'NA'
#         print(cty , state , cntry)
    elif type(a) == str:
        if len(a.split(',')) == 3:
            cty , state , cntry = a.split(',')
#             print(cty , state , cntry)
        elif len(a.split(',')) == 2:
            cty , cntry = a.split(',')
            state = 'NA'
#             print(cty , state , cntry)
        elif len(a.split(',')) == 1:
            cntry = a
            cty , state = 'NA' , 'NA'
#             print(cty , state , cntry)
#     print(cty , state , cntry)
    csc_lst = {
                  'City': cty,
                  'State':state,
                  'Country':cntry
                }
    lst.append(csc_lst)
```

```python
dfi.drop('Country',axis =1,inplace = True)
```

```python
pd.concat([dfi,pd.DataFrame(lst)],axis = 1).to_csv('Notebook Tier Dataset.csv',index = False)
```
