```python
lvl_1 = pd.read_csv('Level 1.csv')
lvl_1.sample(5)
```

```python
lvl_2 = pd.read_csv('Level 2.csv')
lvl_2.sample(5)
```

```python
country = pd.concat([lvl_1,lvl_2],axis = 1)['Country']
country
```

```python
lst = []
for a in country:
    if type(a) == float:
        cty , state , cntry = 'NA' , 'NA' , 'NA'
#         print(cty , state , cntry)
    elif type(a) == str:
        if len(a.split(',')) == 3:
            cty , state , cntry = a.split(',')
#             print(cty , state , cntry)
        elif len(a.split(',')) == 2:
            cty , cntry = a.split(',')
            state = 'NA'
#             print(cty , state , cntry)
        elif len(a.split(',')) == 1:
            cntry = a
            cty , state = 'NA' , 'NA'
#             print(cty , state , cntry)
#     print(cty , state , cntry)
    csc_lst = {
                  'City': cty,
                  'State':state,
                  'Country':cntry
                }
    lst.append(csc_lst)
```

```python
pd.DataFrame(lst).to_csv('city,state,country.csv',index = False)
```
