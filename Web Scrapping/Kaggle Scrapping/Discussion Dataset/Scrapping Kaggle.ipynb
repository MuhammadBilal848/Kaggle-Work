```python
for i in range(2,2856): # 2834
    entry = driver.find_element(By.XPATH, f'//*[@id="site-content"]/div[2]/div/div[2]/div/div[1]/div[2]/div[2]/div/div[{i}]/div/div')
    rank = entry.find_element(By.XPATH, f'//*[@id="site-content"]/div[2]/div/div[2]/div/div[1]/div[2]/div[2]/div/div[{i}]/div/div/div[1]').text
    link = entry.find_element(By.XPATH, f'//*[@id="site-content"]/div[2]/div/div[2]/div/div[1]/div[2]/div[2]/div/div[{i}]/div/div/div[3]/a').get_attribute('href')
    gold = entry.find_element(By.XPATH,f'//*[@id="site-content"]/div[2]/div/div[2]/div/div[1]/div[2]/div[2]/div/div[{i}]/div/div/div[5]/div[1]/div[2]/span').text
    silver = entry.find_element(By.XPATH,f'//*[@id="site-content"]/div[2]/div/div[2]/div/div[1]/div[2]/div[2]/div/div[{i}]/div/div/div[5]/div[2]/div[2]/span').text
    bronze = entry.find_element(By.XPATH,f'//*[@id="site-content"]/div[2]/div/div[2]/div/div[1]/div[2]/div[2]/div/div[{i}]/div/div/div[5]/div[3]/div[2]/span').text
    points = entry.find_element(By.XPATH,f'//*[@id="site-content"]/div[2]/div/div[2]/div/div[1]/div[2]/div[2]/div/div[{i}]/div/div/div[6]').text
    joined = entry.find_element(By.XPATH,f'//*[@id="site-content"]/div[2]/div/div[2]/div/div[1]/div[2]/div[2]/div/div[{i}]/div/div/div[4]/p[2]/span').text
    name = entry.find_element(By.XPATH,f'//*[@id="site-content"]/div[2]/div/div[2]/div/div[1]/div[2]/div[2]/div/div[{i}]/div/div/div[4]/p[1]/a').text
    level = entry.find_element(By.XPATH,f'//*[@id="site-content"]/div[2]/div/div[2]/div/div[1]/div[2]/div[2]/div/div[{i}]/div/div/div[2]/img').get_attribute('title')
    dict_k1 = {
               'Name':name,
               'Rank':rank,
               'Level':level,
               'Link':link,
               'Gold':gold,
               'Silver':silver,
               'Bronze':bronze,
               'Points':points,
               'Joined':joined
    }
    notebook_lst.append(dict_k1)
#     break
#     print('Points',entry.find_element(By.XPATH))
```

```python
df = pd.DataFrame(notebook_lst)
```
