# Country ,City, State

```python
lvl_1 = pd.read_csv('Level 1.csv')
lvl_2 = pd.read_csv('level 2.csv')
```

```python
country = pd.concat([lvl_1,lvl_2],axis = 1)#['Country']
country.drop('Country',axis = 1,inplace = True)
country
```

```python
lst = []
for a in country:
    if type(a) == float:
        cty , state , cntry = 'NA' , 'NA' , 'NA'
#         print(cty , state , cntry)
    elif type(a) == str:
        if len(a.split(',')) == 3:
            cty , state , cntry = a.split(',')
#             print(cty , state , cntry)
        elif len(a.split(',')) == 2:
            cty , cntry = a.split(',')
            state = 'NA'
#             print(cty , state , cntry)
        elif len(a.split(',')) == 1:
            cntry = a
            cty , state = 'NA' , 'NA'
#             print(cty , state , cntry)
#     print(cty , state , cntry)
    csc_lst = {
                  'City': cty,
                  'State':state,
                  'Country':cntry
                }
    lst.append(csc_lst)
```

```python
pd.concat([country,pd.DataFrame(lst)],axis=1).to_csv('Discussion Tier Dataset.csv',index = False)
# pd.DataFrame(lst)
```
