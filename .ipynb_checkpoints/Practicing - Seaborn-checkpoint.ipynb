# sns.replot()
* The Seaborn Relational Plot (relplot) allows us to visualise how variables within a dataset relate to each other.
#### seaborn.relplot(*, x=None, y=None, hue=None, size=None, style=None, data=None, row=None, col=None, col_wrap=None, row_order=None, col_order=None, palette=None, hue_order=None, hue_norm=None, sizes=None,size_order=None, size_norm=None, markers=None, dashes=None, style_order=None, legend='auto', kind='scatter', height=5, aspect=1, facet_kws=None, units=None, **kwargs)
https://seaborn.pydata.org/generated/seaborn.relplot.html

```python
sns.relplot('Sales','Profit',data = sales)
```

```python
# here i color coded the graph with categorical variable 'Order Priority'
sns.relplot('Sales','Profit',data = sales,hue='Order Priority')
```

```python
# we can also set different marking/styles for each element of the categorical variable
sns.relplot('Sales','Profit',data = sales,hue='Order Priority',style='Order Priority')
```

```python
# if we want to check the shipping mode of the order priority.
sns.relplot('Sales','Profit',data = sales,hue='Order Priority',style='Ship Mode')
```

```python
# you can assign any number or any array/series of numbers in size param. 
sns.relplot('Sales','Profit',data = sales,size='Discount')
```

```python
# we can also fix the size of the dots
sns.relplot('Sales','Profit',data = sales,size='Discount',sizes=(80,200))
```

```python
# making a line chart
sns.relplot('Order Date','Sales',data = sales,kind='line')
# the plot is mixed up and unable to give insights so we need to clean the order date column
```

```python
# creating a new column having only years or order date that we extacted using data discriptors .dt.year
sales['Year'] = sales['Order Date'].dt.year
```

```python
sales['Year'].unique()
```

```python
sns.relplot('Year','Sales',data = sales,kind='line')
# shaded area is confidence interval and line btw the C.I is mean
```

```python
# we can set C.I to None
sns.relplot('Year','Sales',data = sales,kind='line',ci=None)
```

```python
# we can set C.I to standdart deviation for statistical analysis
sns.relplot('Year','Sales',data = sales,kind='line',ci='sd')
```

```python
# if we want to see how the sales line drop from year to year we can set estimator to none
sns.relplot('Year','Sales',data = sales,kind='line',ci='sd',estimator = None)
```

```python
sns.relplot('Year','Sales',data = sales,kind='line',hue='Order Priority')
```

```python
sns.relplot('Year','Sales',data = sales,kind='line',hue='Ship Mode')
```

```python
sns.relplot('Year','Sales',data = sales,kind='line',hue='Ship Mode',style='Order Priority')
```

```python
sns.relplot('Year','Sales',data = sales,kind='line',hue='Ship Mode',style='Order Priority',ci=None,estimator=None)
# here insight is that "you had highest sales in 2009 when order priority was high and shipping mode was regular air"
```

```python
# if we want to get clear view or want to see each column saperately we can use col param
sns.relplot('Year','Sales',data = sales,kind='line',hue='Ship Mode',col='Order Priority',ci=None,estimator=None)
# here insight is that "you had highest sales in 2009 when order priority was medium and shipping mode was regular air"
```

```python
sns.relplot('Year','Sales',data = sales,kind='line',hue='Ship Mode',col='Order Priority',ci=None,estimator=None,col_wrap=3)
# col_wrap = 3 means show me 3 graphs side by side
```

```python
sns.relplot('Year','Sales',data = sales,kind='line',hue='Ship Mode',col='Region',ci=None,estimator=None,col_wrap=3)
# col_wrap = 3 means show me 3 graphs side by side
```

```python

sns.relplot('Sales','Profit',data = sales,kind='scatter',hue='Ship Mode',col='Customer Segment',row = 'Product Category')
```

# sns.catplot()
* The catplot() method in this module is designed to work with ****CATEGORICAL**** data efficiently.
#### seaborn.catplot(*, x=None, y=None, hue=None, data=None, row=None, col=None, col_wrap=None, estimator=<function mean at 0x7ff320f315e0>, ci=95, n_boot=1000, units=None, seed=None, order=None, hue_order=None, row_order=None, col_order=None, kind='strip', height=5, aspect=1, orient=None, color=None, palette=None, legend=True, legend_out=True,sharex=True, sharey=True, margin_titles=False, facet_kws=None, **kwargs)
https://seaborn.pydata.org/generated/seaborn.catplot.html

```python
sns.catplot('Ship Mode','Sales', data = sales)
```

```python
sns.catplot('Ship Mode','Sales', data = sales,jitter=False)
# jitter param decreases the magnitude of plotted data.
```

```python
# we can also apply query on data
sns.catplot('Ship Mode','Sales', data = sales.query('Profit > 1000'),jitter = False)
```

```python
# we can create swarm plot by using kind param
sns.catplot('Ship Mode','Sales', data = sales.query('Profit > 1000'),kind='swarm')
# the magnitude is thicker as compared to strip(default plot in catplot) 
```

```python
sns.catplot('Product Category','Shipping Cost',data = sales,kind='box')
# box plot shows that shipping cost of office supplies and product category is very low while the furniture's cost is higher. 
# dots above each categoy represents the outliers.
```

```python
sns.catplot('Product Category','Shipping Cost',hue = 'Ship Mode',data = sales,kind='box')
```

```python
# we can sort hue by using hue_order
sns.catplot('Product Category','Shipping Cost',hue = 'Ship Mode',
            hue_order = ['Delivery Truck','Regular Air','Express Air']
            ,data = sales,kind='box')
```

# sns.distplot()
* Seaborn distplot lets you show a histogram with a line on it. This can be shown in all kinds of variations.
* A distplot plots a univariate distribution of observations. 
#### seaborn.distplot(a=None, bins=None, hist=True, kde=True, rug=False, fit=None, hist_kws=None, kde_kws=None, rug_kws=None, fit_kws=None, color=None, vertical=False, norm_hist=False, axlabel=None, label=None, ax=None, x=None)

```python
sns.distplot(sales['Sales'])
# we can see the sales data is right skewed and outliers are at the end of the tail.
```

```python
sns.distplot(sales['Profit'])
```

```python
sns.distplot(sales['Profit'],kde=False)
# kde(kernel density estimator) remove that histogram
```

```python
sns.distplot(sales['Profit'],hist=False)
# hist shows the histogram
```

```python
sns.distplot(sales['Profit'],hist=False,rug=True)
# rug the concentration of the dataset caused by data points
```

# sns.heatmap()
* Heatmap is defined as a graphical representation of data using colors to visualize the value of the matrix. 
* In this, to represent more common values or higher activities brighter colors basically reddish colors are used and to represent less common or activity values, darker colors are preferred. 
#### seaborn.heatmap(data, *, vmin=None, vmax=None, cmap=None, center=None, robust=False, annot=None, fmt='.2g', annot_kws=None, linewidths=0, linecolor='white', cbar=True, cbar_kws=None, cbar_ax=None, square=False, xticklabels='auto', yticklabels='auto', mask=None, ax=None, **kwargs)

```python
s1 = sales['Order Quantity']
s2 = sales['Discount']
```

```python
heat_data = pd.concat([s1,s2],join='inner',axis=1).head(50)
```

```python
sns.heatmap(heat_data.head(20))
```

```python
sns.heatmap(sales.corr())
# here we can see that where the color is darker there is a strong correlation btw them, but it is not a case while using 
# default settings, sns maps larger values to lighter color and greater to lighter by default(you can see it on the line 
# situated at right side of the graph.
```

```python
# for changing the sns map that sets larger values to darker and lower values to lighter color we will use cmap param
sns.heatmap(sales.corr(),cmap='Reds')
# we can see the change on the left 
```

```python
sns.heatmap(sales.corr(),cmap='RdBu') # starts with red ends on blue
```

```python
sns.heatmap(sales.corr(),cmap='RdBu') # starts with red ends on blue
```

```python
# we can see on the right side line that the canter value is not at 0 , we can center it using center param
sns.heatmap(sales.corr(),cmap='RdBu',center=0) # the corr here is not -ve(to  see the diff,
# we have to set SETTING vmin and vmax PARAMS)
```

```python
sns.heatmap(sales.corr(),cmap='RdBu',vmin=-1,vmax=1)
```

```python
sns.heatmap(sales.corr(),cmap='RdBu',vmin=-1,vmax=1,annot=True)
```

```python
# if we want to add line btw the columns n rows of the graph we can use linewidth param
sns.heatmap(sales.corr(),cmap='RdBu',vmin=-1,vmax=1,linewidths=0.5,linecolor='gray')
# we can also change x and y labels of each row n col using xticklabels and yticklabels param
```
