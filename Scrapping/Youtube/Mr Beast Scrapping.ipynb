```python
lst = []
```

```python
for vid in videos:
    try:
        titles    = vid.find_element(By.XPATH, './/*[@id="video-title"]').text
    except:
        titles    = 'NA'
    try:
        tot_views = vid.find_element(By.XPATH, './/*[@id="metadata-line"]/span[1]').text
    except:
        tot_views = 'NA'
    try:
        up_time   = vid.find_element(By.XPATH, './/*[@id="metadata-line"]/span[2]').text
    except:
        up_time   = 'NA'
    try:
        duration  = vid.find_element(By.XPATH, './/a[@id="thumbnail"]').text
    except:
        duration  = 'NA'
    try:
        sub       = vid.find_element(By.XPATH, './/*[@id="video-badges"]/div').text
    except:
        sub       = 'NA'
    try:
        link      = vid.find_element(By.TAG_NAME, 'a').get_attribute('href')
    except:
        link      = 'NA'
    video_df = { 
            'Title':titles,
            'Total Views':tot_views,
            'Upload Time':up_time,
            'Duration':duration,
            'Subtitles':sub,
            'Link':link    
            }
    lst.append(video_df)

    
# titles    = vid.find_element(By.XPATH, './/*[@id="video-title"]').text
# tot_views = vid.find_element(By.XPATH, './/*[@id="metadata-line"]/span[1]').text
# up_time   = vid.find_element(By.XPATH, './/*[@id="metadata-line"]/span[2]').text
# duration  = vid.find_element(By.XPATH, './/a[@id="thumbnail"]').text
# sub       = vid.find_element(By.XPATH, './/*[@id="video-badges"]/div').text
# link      = vid.find_element(By.TAG_NAME, 'a').get_attribute('href')


# titles    = 'NA'
# tot_views = 'NA'
# up_time   = 'NA'
# duration  = 'NA'
# sub       = 'NA'
# link      = 'NA'
```
