```python
import requests
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
baseurl = 'https://www.coursera.org'
```

```python
header = {
    'User-Agents': 'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/47.0.2526.111 Safari/537.36'
}
```

# Scrapping Links

```python
r = requests.get('https://www.coursera.org/certificates/data-science')
soup = BeautifulSoup(r.content, 'lxml')
```

```python
course_list = soup.find_all('div',class_= 'cds-63 css-qsaw8 cds-65 cds-grid-item cds-134')
```

```python
course_link = []
```

```python
for i in course_list:
    for link in i.find_all('a',href = True):
        course_link.append(baseurl + link['href'])
```

# DataFrame

```python
lst = []
```

```python
for i,link in enumerate(course_link):
    if i in [0,1,2,3]:
        pass
    else:
        r = requests.get(link , headers = header)
        soup = BeautifulSoup(r.content, 'lxml')
        try:
            name = np.array(soup.find('div',class_ ='_1bjlpa11 BannerTitle text-xs-left banner-title-container--without-subtitle').find('h1'))[0]
            ratings = np.array(soup.find('span',class_ = '_16ni8zai m-b-0 rating-text number-rating number-rating-expertise'))[0]
            dls = np.array(soup.find('div',class_ = 'ProductGlance').find_all('span')).reshape(-1)
            dls1 = np.array(soup.find('div',class_ = 'ProductGlance').find_all('div',class_ = '_16ni8zai m-b-0')).reshape(-1)
            duration = dls[0].split(' ')[1]
            lang = dls1[5]
            lvl = dls1[3]
            hpw = dls[1].split(' ')[3]
            sub = dls[-1].split(' ')[1:]
            tot_rat = np.array(soup.find('div',class_ = '_wmgtrl9 color-white ratings-count-expertise-style').find_all('span')).reshape(-1)[1].split(' ')[0]
            enroll = np.array(soup.find('div',class_ = '_1fpiay2').find('span').find('span'))[0]
            p_ex = 'NA'
        except:
            name = 'NA'
            ratings = 'NA'
            duration = 'NA'
            lang = 'NA'
            lvl = 'NA'
            hpw = 'NA'
            sub = 'NA'
            tot_rat = 'NA'
            enroll = 'NA'
            p_ex = 'NA'
        course = {
                  'Name':name,
                  'Ratings': ratings,
                  'Duration': duration,
                  'Language': lang, 
                  'Level': lvl,
                  'Total Ratings': tot_rat,
                  'Enrolled': enroll,
                  'Hours Per Week': hpw, 
                  'Subtitles': sub,
                  'Prerequisite': p_ex,
                  'Link': link
        }
        lst.append(course)
```
